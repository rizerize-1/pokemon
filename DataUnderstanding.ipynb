### **Data Visualization and Storytelling - Pokemon Dataset**

- Dữ liệu Pokemon được lấy từ cuộc thi Kaggle ["Pokemon - Weedle's Cave"](https://www.kaggle.com/competitions/pokemon-weedles-cave/data) và ["The Complete Pokemon Dataset"](https://www.kaggle.com/datasets/rounakbanik/pokemon/data) từ Kaggle
- Dữ liệu này chứa thông tin cơ bản về các loài Pokemon, bao gồm các thuộc tính như tên, loại, chỉ số chiến đấu và các đặc điểm khác. Mục đích ban đầu của dataset Weedle's Cave là để dự đoán kết quả của các trận đấu 2 Pokemon bất kỳ dựa trên các thuộc tính của chúng, dataset còn lại là để cung cấp thêm các thông tin chi tiết về đặc điểm của các Pokemon được tổng hợp từ [trang web này](https://serebii.net/pokemon) và được bổ sung bởi [Daalma7](https://github.com/Daalma7) từ thông tin trên [bulbapedia](https://bulbapedia.bulbagarden.net/)
</span>

<span style = "font-family: Verdana; font-size: 20px">

#### **I. Các file được sử dụng trong dự án**
</span>

<span style = "font-family: Verdana; font-size: 20px">

##### **1. `combats.csv`**
- **Nội dung:** Chứa thông tin về các trận đấu giữa các Pokemon, bao gồm ID của hai Pokemon tham gia và kết quả của trận đấu
- **Kích thước:** Dữ liệu bao gồm 50000 dòng và 3 cột
- **Mục đích sử dụng:** Dùng để đánh giá hiệu quả của quá trình chuẩn bị dữ liệu và làm input cho mô hình dự đoán kết quả trận đấu giữa hai Pokemon
<span>

<span style = "font-family: Verdana; font-size: 20px">

##### **2. `pokemon.csv`**
- **Nội dung:** Chứa các thông tin cơ bản về các loài Pokemon, bao gồm tên, loại, chỉ số chiến đấu và các đặc điểm khác
- **Kích thước:** Dữ liệu bao gồm 800 dòng và 12 cột
- **Mục đích sử dụng:** Dùng để thêm features cho dataset combats thông qua `merge` trên cột `#`

<span style = "font-family: Verdana; font-size: 20px">

##### **3. `pokemonfullstats.csv`**
- **Nội dung:** Chứa các thông tin cơ bản về các loài Pokemon gần giống như `pokemon.csv`, nhưng có thêm các chỉ số chi tiết hơn về đặc điểm vật lý, các tương tác giữa các thuộc tính và kỹ năng đặc biệt của Pokemon
- **Kích thước:** Dữ liệu bao gồm 1179 dòng và 46 cột
- **Mục đích sử dụng:** Dùng để thêm features cho dataset pokemon thông qua `merge` trên cột `Name` (Vấn đề này sẽ được giải thích kỹ hơn ở dưới)
